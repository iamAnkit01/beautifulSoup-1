```python
#Q1;  1) Write a python program to display all the header tags from wikipedia.org. 
#importing all the imporatant libaries for scrapping
from bs4 import BeautifulSoup
import requests
#requesting the web page to get access form their hosting
page=requests.get('https://www.wikipedia.org/') 
#Checking the response code wether the hosting allows scrapping or not
page
# the site allows scarpping data to beautiful soup first we downlad the content of the web page content
content= BeautifulSoup(page.content)
#scraping h1 tag of that place  in a variable hearder using for loop to find all the possible h1 tags on the page
header=[]
for i in content.find_all('span',class_="central-textlogo__image sprite svg-Wikipedia_wordmark"):
    header.append(i.text.replace('\n',''))
for i in content.find_all('h2',class_="bookshelf-container"):
    header.append(i.text.replace('\n',''))
#all header tag scrapped
header    
```

```python
#2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame
# requesting imdb hosting to sccrap the data from the webpage
imdb= requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
imdb
#downloading the web content through BeautifulSoup
web= BeautifulSoup(imdb.content)
movie_name=[]
#scrapping movie name
for i in web.find_all('h3',class_="lister-item-header"):
    for j in i.find_all("a"):
        movie_name.append(j.text)
# scrapping year of the movies through downoaded web page         
relase=web.find_all('span',class_="lister-item-year text-muted unbold")
year=[]
for i in relase:
    a=i.text.replace('(','')
    year.append(a.replace(')',''))
# scrapping ratings of the movies through downoaded web page
ratings=[]
for i in web.find_all('div',class_="inline-block ratings-imdb-rating"):
    ratings.append(i.text.replace('\n',''))
print(len(movie_name),len(year),len(ratings))
```

```python
# we have to scarp 100 so that we go to the next page url to scrap more data till count 100.
imdb= requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
#repate the above process to get the data 
web= BeautifulSoup(imdb.content)

#scrapping movie name
for i in web.find_all('h3',class_="lister-item-header"):
    for j in i.find_all("a"):
        movie_name.append(j.text)

relase=web.find_all('span',class_="lister-item-year text-muted unbold")
for i in relase:
    a=i.text.replace('(','')
    year.append(a.replace(')',''))
    
for i in web.find_all('div',class_="inline-block ratings-imdb-rating"):
    ratings.append(i.text.replace('\n','')) 
print(len(movie_name),len(year),len(ratings))    
```

```python
# creating dataframe of scarpped data
import pandas as pd
df=pd.DataFrame({('Movie Name'):movie_name,('Release Year'):year,('IMDb ratings'):ratings})
df
```

```python
#Q: 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame. 
web1= requests.get('https://www.imdb.com/list/ls009997493/')

content= BeautifulSoup(web1.content)

#scraping movie name from the downloaded web page
move_name=[]
for i in content.find_all('h3',class_="lister-item-header"):
    for x in i.find_all('a'):
        move_name.append(x.text)
    
relase=content.find_all('span',class_="lister-item-year text-muted unbold")
year=[]
for i in relase:
    a=i.text.replace('(','')
    year.append(a.replace(')',''))
    
# scrapping ratings of the movies through downoaded web page
ratings=[]
for i in content.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text.replace('\n',''))  
    
print(len(move_name),len(year),len(ratings))    
```

```python
#Q: Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp . 
```

```python
#requesting hosting server to get persmission to scrap the web page 
messo=requests.get('https://meesho.com/bagsladies/pl/p7vbp')

#downloading web page content
meesho=BeautifulSoup(messo.content)

#Scrapping the product name with class name
product=[]
for i in meesho.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    product.append(i.text)
    
price=[]
for i in meesho.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    price.append(i.text)
    
discount=[]
for i in meesho.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
    discount.append(i.text)
    
print(len(product),len(price),len(discount))      
```

```python
# creating dataframe of scarpped data
df=pd.DataFrame({('Product'):product,('Price'):price,('Discount'):discount})
df
```
